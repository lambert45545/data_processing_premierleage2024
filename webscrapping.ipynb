```python
import codecs
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd

f = codecs.open("Premier.html", 'r', 'utf-8')
soup = BeautifulSoup(f.read(), "lxml")
players = soup.find_all("a", {"class": "player__name"}, href=True)
players_info = []

for player in players:
    href = player['href']
    href_stats = href.replace("overview", "stats")

    page1 = requests.get(href)
    src = page1.content
    soup1 = BeautifulSoup(src, "lxml")
    club = soup1.find("span", {"class": "player-club-history__team-name player-club-history__team-name--long"})
    country = soup1.find("span", {"class": "player-overview__player-country"})

    other_info = soup1.find("section", {"class": "player-overview__side-widget"})
    if other_info is not None:
        info = other_info.find_all("div", {"class": "player-overview__info"})

    data = {}

    if club and info is not None and len(info) == 5:
        try:
            data["player"] = f"{soup1.find('div', {'class': 'player-header__name-first'}).text.strip()} " \
                             f"{soup1.find('div', {'class': 'player-header__name-last'}).text.strip()}"
            data["country"] = country.text.strip()
            data["age"] = info[1].text.strip()
            data["Position"] = info[4].text.strip()
            data["height"] = info[2].text.strip()
            data["club"] = club.text.strip()

            page2 = requests.get(href_stats)
            src2 = page2.content
            soup2 = BeautifulSoup(src2, "lxml")

            stats = soup2.find_all("span", {"class": "js-all-stat-container"})

            # Check if all stats are present before appending the data

            for stat in stats:
                data[stat.get('data-stat')] = stat.text.strip()

            players_info.append(data)

        except AttributeError as e:
            print(f"Error processing player: {e}")

# Writing to CSV
keys=[]
for player_info in players_info:
    for key in player_info.keys():
        if key not in keys:
            keys.append(key)

with open("players.csv","w") as OF:
          dict_writer=csv.DictWriter(OF,keys)
          dict_writer.writeheader()
          dict_writer.writerows(players_info)
          print("file created")
print(len(players_info))



# Writing CSV Header
```
